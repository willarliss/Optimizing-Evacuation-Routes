# Pull Data From the HERE API

Resources:
- https://developer.here.com/documentation/examples/rest/traffic/traffic-incidents
- https://developer.here.com/documentation/traffic/dev_guide/topics/resource-parameters-incidents-6.1.html

```python
#create a bbox (geographic boundaries for api to pull information)
bbox = 'bbox=30.7218%2C-94.9996%3B29.4288%2C-91.4400'
```

<img src="./imgs/bbox.png">

```python
# Create url for API call.
base_url = 'https://traffic.ls.hereapi.com/traffic/6.3/incidents.json?' + bbox  #setting bounding box to lousiana
```

```python
api_key = 'apiKey=<insert api key here>' #commented out for privacy reasons
```

```python
extra_params = 'c=US&lg=en&i18n=true&localtime=true&units=metric&startTime=2020-08-20' #set a start time of 8/20 
```

### Create the request

```python
res = requests.get(base_url + '&' + api_key + '&' + extra_params)

#check the status
res
```

```python
# Convert to json
traffic = res.json()
```

```python
#create a dataframe
traffic_df = pd.DataFrame(traffic['TRAFFIC_ITEMS']['TRAFFIC_ITEM'])
```

```python
#explore the data
traffic_df.shape
```

#### View as a Dataframe

```python
traffic_df.head()
```

## Clean the columns Data

#### Criticality column

```python
#create a column with how critical the traffic item is

traffic_df['CRITICAL_VALS'] = [row['DESCRIPTION'] for row in traffic_df['CRITICALITY']]
```

```python
traffic_df.head(2)
```

#### Abbreviation column

```python
incident = []

for row in traffic_df['ABBREVIATION']:
    if row['DESCRIPTION'] == '':
        incident.append('NA') # need to account for the blank values
    else:
        incident.append(row['DESCRIPTION'])
        
# add description values as a new column in the data frame
traffic_df['INCIDENT_TYPE'] = incident
```

```python
traffic_df.head(2)
```

#### Location column

```python
## Create a dataframe to break up the LOCATION column
location = pd.DataFrame(list(traffic_df['LOCATION']))
location.head(3)
```

```python
#get the latitude and longitude of the origin geolocation
geoloc_start = []

for row in location['GEOLOC']:
        geoloc_start.append(row['ORIGIN'])
        
#create a dataframe        
geoloc_start_df = pd.DataFrame(geoloc_start)

#rename the columns
geoloc_start_df.columns = ["LATITUDE_START","LONGITUDE_START"]

geoloc_start_df.head(2)
```

```python
# Get the endpoint of the longitude and latitude
geoloc_end = []

for row in location['GEOLOC']:
        geoloc_end.append(row['TO'][0])
        
#create a dataframe        
geoloc_end_df = pd.DataFrame(geoloc_end)

#rename the columns
geoloc_end_df.columns = ["LATITUDE_END","LONGITUDE_END"]

geoloc_end_df.head(2)
```

### MERGE BOTH WITH THE ORIGINAL DATAFRAME

```python
traffic_df = pd.merge(traffic_df,geoloc_start_df, how="left", left_index=True, right_index=True)
traffic_df = pd.merge(traffic_df,geoloc_end_df, how="left", left_index=True, right_index=True)

traffic_df.head(2)
```

#### Traffic Detail Column

```python
#Add a column showing if the road was closed or not

road_closure = []

for row in traffic_df['TRAFFIC_ITEM_DETAIL']:
    road_closure.append(row['ROAD_CLOSED'])
    
#add this a new column to the dataframe
traffic_df['ROAD_CLOSED'] = road_closure

traffic_df.head(2)
```

#### Traffic Description Column

```python
traffic_desc = []

for row in traffic_df['TRAFFIC_ITEM_DESCRIPTION']:
    traffic_desc.append(row[1]['value'])
```

```python
#clean up errors with commas
traffic_desc = list(map(lambda item : item.replace(', ',''), traffic_desc))
traffic_desc[:4]
```

```python
#Add to our dataframe

traffic_df['TRAFFIC_DESC'] = traffic_desc

traffic_df.head(3)
```

## Create a new dataframe to save as a csv 

```python
clean_traffic = traffic_df[['TRAFFIC_ITEM_ID', 'ORIGINAL_TRAFFIC_ITEM_ID',
       'TRAFFIC_ITEM_STATUS_SHORT_DESC', 'TRAFFIC_ITEM_TYPE_DESC',
       'START_TIME', 'END_TIME', 'ENTRY_TIME', 'VERIFIED','CRITICAL_VALS', 'INCIDENT_TYPE', 'LATITUDE_START', 'LONGITUDE_START',
       'LATITUDE_END', 'LONGITUDE_END', 'ROAD_CLOSED', 'TRAFFIC_DESC']]

clean_traffic.head()
```

```python
#Save as a csv
clean_traffic.to_csv('./GeoCoding Data/here_api_data.csv',index=False)
```
